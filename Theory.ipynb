## ${\textbf{Hierarchical Clustering}}$

* What we have is Agglomerative HC (bottom up approach). 

> 1. First all 2000 points (each having 7 field standardized values) are considered as individual clusters. 

> 2. Then we find a cluster of 2 points (here mean of all 7 fields considered as point value) with minimum distance (based on euclidean measure) . Hence having 1000 clusters. 

> 3. Finally, the 1000 pairs's means (refered to as centroids) is found and then step 2 is repeated again forming 500 clusters and so on until 1 super cluster is found

* Ward is finally applied to the dendogram to obtain final number of clusters (4 in our case). Which works as below

> Objective is to find a horizontal line (a distance threshold) at which to make the cut and obtain the # of clusters from below it.

> Rule: Find the longest vertical line which isn't intersected by any horizontal lines in the dendogram. The root of such line is where to make the cut. (refer pic)

## Cross Price Elasticity Brand 5, Cross Brand 4

### Interpreting the graph 

#### Elasticity is all +ve. It indicates the elasticity when the price values of the competetor brand increases. (i.e) 
#### When our comp inc. their brand price, our brand prob(buy) inc.

#### If cross price E > 0, then the 2 products are considered substitutes. (therefore, brand 4 could take over brand 5 if price inc. is hiked)

#### If mod[E(cross brand)] > mod[E(own brand)], then the other/comp pdt is a strong substitute. 

> (e.g) For price range b/w 0 to 1.65
>> E[cross brand] ranges from [0.33 to 0.55] ---(1) and

>> E[own brand] ranges from [0 to 0.51] --- (2)

>> We can see that (1) > (2) and hence brand 4 is a strong substitue. 

>> BUT!! When we look at the min and max prices of brand 4 (the selling price range), we see that 0 - 1.65 doesn't overlap with selling price. (i.e) brand 4 would be a great substitute if they reduced their prices. But in the selling price range, we can see that (1) < (2) and hence brand 4 is weak substitute!

>> So a marketing stratergy might be that we can promote brand 5 to brand 4 choosers since b5 is strong substitute for b4 in the b4 selling price range

## Own and Cross-Price Elasticity by Segment
